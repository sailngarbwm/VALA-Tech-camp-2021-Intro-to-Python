```python
file = open("Bohemian_rhapsody.txt","r")
file_all = file.read()

print(file_all)
```

```python
file_all
```

```python
file_all = file_all.replace('\n',' ')
file_all = file_all.lower()
```

```python
no_list = [',','?','.','(',')','!','-']

for no in no_list:
    if no in file_all:
        file_all = file_all.replace(no,' ')
```

```python
boheme_list = file_all.split(' ')
```

```python
boheme_list
```

```python
while ('' in boheme_list) or (' ' in boheme_list):
    if "" in boheme_list:
        boheme_list.remove('')
    if " "in boheme_list:
        boheme_list.remove('')
```

```python
boheme_list
```

```python
import pandas as pd
import numpy as np
boheme_dict = {'words':np.array(boheme_list)}
boheme_series = pd.Series(data = boheme_list, dtype=str)
```
